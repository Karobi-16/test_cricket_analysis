# Description of the columns


1. **Player:**     Name of the player (and country/team name - he has played for)

2. **Span:**       Test cricket careers of the player (played for the team)

3. **Mat:**        Total number of matches he played

4. **Inns:**       Total number of innings he played

5. **Balls:**         Total number of balls the player bowled
6. **Runs:**      total  runs given by the player
13. **BBI:**       BBI stands for Best Bowling in Innings and only gives the score for one innings.

14. **BBM:**       BBM stands for Best Bowling in Match and gives the combined score over all innings

7. **Wkts:**       Total number of Wickets taken by the player

8. **Ave:**        Average runs given by the player( per wicket)

9. **Econ:**        Average runs given by the player( per over)

10. **SR:**         Average number of balls taken by the player(per Wickets)

11. **5:**         Total number of innigs bowler took at least 5 wickets
12. **10:**        Total number of matches bowler took at least 10 wickets

#### Remove the columns BBI and BBM.

```python
df= df.drop('Best_Bowling_Matches', axis=1)
display(df.head(3))
```

```python

file = df.copy()

#extracting the country information from the 'Player' column 

file['Player'] = file['Player'].str.replace("(", "")
file['Player'] = file['Player'].str.replace(")", "")
file['Player'] = file['Player'].str.replace("\xa0", " ")
file_player = file['Player'].str.split(" ")
file_player = pd.DataFrame(file_player)

file_player['Country'] = file_player['Player'].str[-1]

print(type(file_player))


display(file_player.head(3))
```

#### Adding new Country column to the data frame and player name in player column

```python
# splitting the 'Player' column to get the information about 'Country'
df_player = df['Player'].str.split("(", expand=True)

# concatinating 'Country' with the main dataframe
df = pd.concat([df, df_player], axis=1)

# dropping the 'Player' columns
df = df.drop('Player', axis=1)

# renaming the column names
df = df.rename(columns={0: 'Player',
                        1: 'Country'})

# remove the ")" from the 'Country' column
df['Country'] = df['Country'].str.replace(")", "")


                                                
                        
# rearrange the columns
new_col_sequence = ['Player', 'Country', 'Span', 'Match', 'Innings', 'total_balls', 'Total_runs', 'Wickets',
                    'Average', 'Economy', 'Strike_Rate', '5_Wickets', '10_wickets']
df = df[new_col_sequence]

display(df.head(3))
```

#### How many players played for ICC?

```python

def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
    
 
df['played_for_ICC'] = df['Country'].apply(icc_check)
df['played_for_ICC'].value_counts()
```

5 players played for ICC


#### How many different countries are present in this dataset?

```python
df['Country'].value_counts()
```

10 individual countries and 4 countries under ICC

```python
# splitting the 'Span' column based on the "-"
df_span = df['Span'].str.split("-", expand=True)

# conccatinating the new dataframe with the main dataframe
df = pd.concat([df, df_span], axis=1)

# renaming the newly created column names
df = df.rename(columns={0: "start_year",
                        1: "end_year"})

# removing the "Span" column
df = df.drop("Span", axis=1)

display(df.head(3))
```

```python
df.info()
```

```python
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')

df['period'] = df['end_year'] - df['start_year']

df = df.drop(["start_year", "end_year"], axis=1)
display(df.head(3))
```
