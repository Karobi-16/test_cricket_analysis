# Description of the columns


1. **Player:**     Name of the player (and country/team name - he has played for)

2. **Span:**       Test cricket careers of the player (played for the team)

3. **Mat:**        Total number of matches he played

4. **Inns:**       Total number of innings he played

5. **NO:**         Total number of matches the player remain not out

6. **Runs:**       Total runs of the player of his test career

7. **HS:**         Highest score of his test career

8. **Ave:**        Average run of his total career

9. **100:**        number of  total centuries of a player

10. **50:**        number of  half-centuries of a player

11. **0:**         number of  total ducks(dismissal with a score of zero) of a player

### Rename the column names appropriately

```python
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'NO': 'Not_out',
                        'Runs': 'Total_runs',
                        'HS': 'Highest_score',
                        'Ave': 'Average',
                         '100': 'Centuries',
                         '50': 'Half_centuries',
                         0: 'Ducks'})

display(df.head(3))
```

### Remove a column

```python
df= df.drop('Not_out', axis=1)
display(df.head(3))
```
